## 1a) Image and Filter Dimensions

### 🖼️ Image  
**Q:** What are the values of \( N \), \( C_{in} \), \( H \), and \( W \) for the image?  

**A:**  
- **\( N \) (Batch size) = 1** → Since `.reshape(1, 1, 512, 512)`, the image is treated as a single batch.  
- **\( C_{in} \) (Number of input channels) = 1** → The image is grayscale, so it has only one channel.  
- **\( H \) (Height) = 512** → The image has a height of **512 pixels**.  
- **\( W \) (Width) = 512** → The image has a width of **512 pixels**.

### 🎛️ Filter  
**Q:** What are the values of \( C_{out} \), \( C_{in} \), \( F_H \), and \( F_W \) for the filter?  

**A:**  
- **\( C_{out} \) (Number of output channels) = 2** → The filter set contains **two filters**.  
- **\( C_{in} \) (Number of input channels) = 1** → Each filter operates on a single grayscale channel.  
- **\( F_H \) (Filter height) = 3** → The height of each filter is **3 pixels**.  
- **\( F_W \) (Filter width) = 3** → The width of each filter is **3 pixels**.  


### Padding
**Q:** If we assume a stride of 1, how must the image x be zero-padded so that the filtered image will have the same spatial resolution as the original image?


**A:**
- Formula for padding \( Padding = ( \( F_H \) - 1 ) / 2 )
- For a 3x3 filter in our case: \( Padding = ( \( 3 \) - 1 ) / 2 ) = 1
- Need **1 pixel** of padding on each side of the image
